```python
import numpy as  np
import matplotlib.pyplot as plt

f = lambda  x: (1 - x)**2 + 100*(x - x**2)**2 # Rosenbrock function
x = np.linspace(-5,10,200)
plt.plot(x,f(x))
plt.xlabel("x")
plt.ylabel("f(x)")

f_p = lambda x: -2*(1 - x) + 200*x - 600*x**2 + 400*x**3
x = np.linspace(-5,10,200)
plt.plot(x,f_p(x))
plt.plot(x,x*0,'k--')
plt.xlabel("x")
plt.ylabel("f_p(x)")

f_pp = lambda x: 202 - 1200*x + 1200*x**2 
x0 = 2
x1 = x0 - f_p(x0)/f_pp(x0)
print(x1)
x2 = x1 - f_p(x1)/f_pp(x1)
print(x2)
x3 = x2 - f_p(x2)/f_pp(x2)
print(x3)

# Newtonian metod
def newton(f, f_p, f_pp, x0, xtol = 0.001):    
    deltaX = xtol * 2 
    x = x0 
    while np.abs(deltaX) > xtol: 
        deltaX = f_p(x)/f_pp(x)
        x = x - deltaX
        # print(x)
    return x 
found_min = newton(f,f_p,f_pp,x0, xtol = 1e-10)

print("Found Minimum: ", found_min)

#Gradient Method
x = np.linspace(-5,10,200)
f_p = lambda x: -2*(1 - x) + 200*x - 600*x**2 + 400*x**3
gamma = 0.01
x0 = 2
x1 = x0 - gamma * f_p(x0)
print(x1)
x2 = x1 - gamma * f_p(x1)
print(x2)
x3 = x2 - gamma * f_p(x2)
print(x3)

def gradient_descent(f_p,x0, xtol=1e-3, gamma = 1e-3):
    deltaX = xtol * 2 
    x = x0 
    iterations = 0
    while np.abs(deltaX) > xtol:
        iterations = iterations +1
        deltaX = gamma * f_p(x)
        x = x - deltaX
        # print(x)
    print(iterations)
    return x 
    
found_min = gradient_descent(f_p, x0, xtol = 1e-7, gamma = 1e-3)

print("Found Minimum: ", found_min)
```
